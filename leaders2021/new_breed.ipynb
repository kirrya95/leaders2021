```python
def load_dataset(data_path, is_train):
  if is_train:
      train_dataset = torchvision.datasets.ImageFolder(
          root=data_path,
          transform=transforms.Compose([
                          transforms.ToTensor(),
                          # transforms.RandomHorizontalFlip(),
                          transforms.Resize((80,80)),
                          transforms.Normalize([0.485, 0.456, 0.406],
                              [0.229, 0.224, 0.225]) 

          ])
      )
  else:
      train_dataset = torchvision.datasets.ImageFolder(
          root=data_path,
          transform=transforms.Compose([
                          transforms.ToTensor(),
                          # transforms.RandomHorizontalFlip(),
                          transforms.Resize((80,80)),
                          transforms.Normalize([0.485, 0.456, 0.406],
                              [0.229, 0.224, 0.225]) 

          ])
      )
  return train_dataset

def get_loader(train_dataset):
    train_loader = torch.utils.data.DataLoader(
        train_dataset,
        batch_size=64,
        num_workers=0,
        shuffle=True
    )
    return train_loader
```

```python
torch.manual_seed(42)
dataset = load_dataset('Images', is_train=True)
print(len(dataset))
```

```python
torch.manual_seed(42)
train_data, val_data = torch.utils.data.random_split(dataset, (6100, 1311))
train_loader = get_loader(train_data)
val_loader = get_loader(val_data)
```

```python
def get_top_3(x):
  #returns top 3 classes
  sort_index = np.argsort(x)
  return(sort_index[-3:][::-1])
```

```python
class NewModel(nn.Module):
  def __init__(self, n_classes, model_start):
    super().__init__()

    self.first_layer = model_start

    self.n_in = 1000
    self.n_classes = n_classes
    self.last_layer =nn.Sequential(
                      nn.Linear(self.n_in, 1024),
                      nn.Tanh(),
                      nn.Linear(1024, 256),
                      nn.Tanh(),
                      nn.Linear(256, 64),
                      nn.ReLU(),
                      nn.Linear(64, 44)
                      )

  def forward(self, x):
    return (self.last_layer(self.first_layer(x)))
```

```python
def training(model, trian_dataloader, optimizer, criterion, device):
  err = 0
  pbar = tqdm(train_loader, desc=f"loss={err}")
  model.eval()
  for (data, target) in pbar:
    data = data.to(device)
    target = target.to(device)
    optimizer.zero_grad()
    prediction = model(data)
    loss = criterion(prediction, target)
    loss.backward()
    optimizer.step()
    err = loss.to('cpu').item()
    pbar.set_description(f"loss: {err:.4}")


def testing(model, val_dataloader, device):
  acc = 0
  accuracy = []
  pbar = tqdm(val_loader, desc=f"acc={acc}")
  model.eval()
  for (data, target) in pbar:
    data = data.to(device)
    target = target.to(device)
    predict = model(data).to('cpu').detach().numpy()
    ans = 0
    k = 0
    for (arr, targ) in zip(predict, target.to('cpu').detach().numpy()):
      ind = get_top_3(arr)
      ans += sum(ind == targ)
      k += 1
    accuracy.append(ans / k)
    pbar.set_description(f"loss: {ans/k}")
  print(f"accuracy: {np.mean(accuracy)}")
  return(np.mean(accuracy))
```

```python
model_start = torchvision.models.vgg16(pretrained=True)
model_start.add_module()
```

```python
model = torchvision.models.resnet101(pretrained=True)
for param in model.parameters():
  param.requires_grad=False
model.fc = nn.Sequential(
                      nn.Linear(2048, 1024),
                      nn.Dropout(0.2),
                      nn.Tanh(),
                      nn.Linear(1024, 512),
                      nn.Dropout(0.2),
                      nn.Tanh(),
                      nn.Linear(512, 256),
                      nn.ReLU(),
                      nn.Linear(256, 44)
                      )

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-4)
```

```python
print(model.parameters)
```

```python
num_epochs = 50
best_er = 0
model.to('cuda')
for e in range(num_epochs):
  training(model, train_loader, optimizer, criterion, 'cuda')
  er = testing(model, val_loader, 'cuda')
  if (er > best_er):
    torch.save(model.state_dict(), 'model_2.pt')
    best_er = er
```
